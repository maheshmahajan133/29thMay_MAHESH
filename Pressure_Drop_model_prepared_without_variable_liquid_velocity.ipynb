```python
import math 
import cmath
g=9.81                 #m/s2
temp=input("Enter the temperature in degree celcius for air:")
t_air=float(temp)
T=273.15+t_air
den_g=10**(-5)*t_air**2-0.0044*t_air+1.2877                     #Kg/m3
print("Density of the air is :",den_g)
visco_g=-9*pow(10,-8)*t_air**2+0.0002*t_air+0.0623                  #Pa.hr
visco_g=visco_g/3600                #Pa.s
print("Dynamic Viscosity of air is in Kg/m.s :",visco_g)
Dg=diffu_g=5*pow(10,-7)*t_air**2+0.0005*t_air+0.0684
Dg=Dg/3600              #
print("Air diffusivity is euqal to :",Dg)
at=float(input("Enter the packing density of the dehumidifier in m2/m3:"))               #packing density  m2/m3
volume_of_tank=0.6*0.6*0.75    #volume of the tank of dehumidifier in m3
thickness_sheet=float(input("Enter the thickness of the Sheet in m:"))    #consider the thickness of the sheet in m
void_fraction=(volume_of_tank-at*0.5*volume_of_tank*thickness_sheet)/(volume_of_tank)                #for 256 it comes out to be 0.872
print("void_fraction:",void_fraction)
Dp=6*(1-void_fraction)/at                                                             # Equivalent diameter of flow channel in meter
print("Dp=",Dp)

desiccant_flow=input("Enter the Desiccant flow rate in Kg/m2.s :")     #Desiccant flow rate in Kg/m2.hr          #proper value need to be eneterd i guesss as in hour
ratio=input("Enter the L/G ratio :")                                  #Liquid to air flow rate 
ratio=float(ratio)
L=float(desiccant_flow)
G=L/ratio

u_l=0.02     #m/s 


# to calculate the actual available area of sheet in z directuon which is used for air circulation
length_sheet=float(input("Enter the length of sheet in meter:"))
height_sheet=float(input("Enter the height of sheet in meter:"))
no_sheet=at*0.5*volume_of_tank/(length_sheet*height_sheet)
area_sheet_airflow=0.6*0.75-length_sheet*thickness_sheet*no_sheet
air_velocity=G/(area_sheet_airflow*den_g)
print("Air velocity m/s:",air_velocity) 

#  to calculate the superficial velocity 
u_g=void_fraction*air_velocity
print("Superficial velocity of the air is :", u_g)

#DRY PRESSURE DROP for Mellapack250Y
m1=1
m2=1
m3=0.32
reg=u_g*den_g*Dp/(visco_g)
print("Reynolds number for given system for air",reg)
f_0=m1/reg+m2/reg**0.5+m3
delta_Pd=0.125*f_0*den_g*u_g**2*at/void_fraction**4.65
print("The dry bed pressure drop in Pa/m:",delta_Pd)

#To calculate the surface tension of solution
temp_desiccant=input("Enter the desiccant inlet temperature in degree celcius:")
t_des=float(temp_desiccant)

conc3=input("Enter the Concentration of the solution:")
x=float(conc3)

surface_tension_0water=0.2358  # N/m
zeta= -0.625
mu= 1.256
theta2=((t_des+273.15)/647.3)
surface_tension_water=surface_tension_0water*(1-zeta*(1-theta2))*((1-theta2)**mu)
print('the value of the surface tension of water at that temperature is {}  mN/m'.format(surface_tension_water))
b1=2.757115
b2=-12.011299
b3=14.751818
b4=2.443204
b5=-3.147739
surface_tension_solution= surface_tension_water * (1+(b1*x)+(b2*theta2*x)+(b3*x*theta2**2)+(b4*x**2)+(b5*x**3))  #N/m
print('the value of the surface tensio of solution at that temperature is {}  N/m'.format(surface_tension_solution))

#To calculate the density of LiCl solution
import math
c1=1.9937718430
c2=1.0985211604
c3=-0.5094492996
c4=-1.7619124270
c5=-44.9005480267
c6=-723692.2618632
thetha3=(t_des+273.15)/647.3
tau=1-thetha3
y=(x/(1-x))
critical_density_of_water=322   # kg/m3
density_of_water=round(critical_density_of_water*(1+c1*tau**(1/3)+c2*tau**(2/3)+c3*tau**(5/3)+c4*tau**(16/3)+c5*tau**(43/3)+c6*tau**(110/3)),4)
print ("The density of the water at given temperature is {}" .format(density_of_water))
f1=1.0
f2=0.540966
f3=-0.303792
f4=0.100791
density_of_solution=round(density_of_water*(f1+f2*y**2+f3*y**3+f4*y**4),4)
print ("the density of the LiCl soltution at {} celcius, {} concentration is {} Kg/M3 ".format(t_des,x,density_of_solution))
rho_l=density_of_solution
print(rho_l)

#To calculate the viscosity of the solution
c1=1.0261862
c2=12481.702
c3=-19510.923
c4=7065.286
c5=-395.561
c6=143922.996
visco_water_0celci=0.001793 # Pa.second
thetha_visco=(t_des/228)-1
print(thetha_visco)
visco_water_temp=(2*10**-11*t_des**4-6*10**-9*t_des**3+7*10**-7*t_des**2-5*10**-5*t_des+0.0017)      # Pa.s 
print(visco_water_temp)
d1=0.090481
d2=1.390262
d3=0.675875
d4=-0.583517
falta=(x/(1-x)**(1/0.6))
print(falta)
visco_solu= visco_water_temp * (math.exp((d1*falta**3.6+d2*falta+d3*(falta/thetha_visco)+d4*falta**2)))      #Pa.s converted to Pa.hr
print ('The value of the solution viscosity is {} in Pa.s'.format(visco_solu))

#operating holdup
h_dyno_0=3.6*(u_l*at**0.5/g**0.5)**0.66*(visco_solu*at**1.5/(rho_l*g**0.5))**0.25*(surface_tension_solution*at**2/(rho_l*g))**0.1
print("Operating Liqiid Holdup hb:",h_dyno_0)

#to calculate the liquid drop diameter 
#for random packing c=0.4 if not for structured packing c=0.8
c_l=0.8
d_l=c_l*(6*surface_tension_solution/((rho_l-den_g)*g))**0.5
print("Diameter of the liquid drop in m:",d_l)


delta_P_tot=delta_Pd
z=0
error=1
while error >0.0000001:
  h_dyno=h_dyno_0*(1+(6*delta_P_tot/(rho_l*g))**2)            #1
  a_l=6*h_dyno/d_l                                       #2
  delta_P_tot1=delta_Pd*((a_l+at)/at)*(void_fraction/(void_fraction-h_dyno))**4.65 #3    
  error=abs(delta_P_tot1-delta_P_tot)
  delta_P_tot=delta_P_tot1
  z=z+1
  print("{}  total pressure drop is {} with error{} and Dynamic Holdup below the loading point {} and Dynamic holdup {} and specific liquid surface area,m2/m3 {}".format(z,delta_P_tot, error,h_dyno_0,h_dyno,a_l))
print("specific liquid surface area,m2/m3:",a_l)
print("Dynamic Holdup below the loading point:",h_dyno_0)
print("Dynamic holdup:",h_dyno)


#pressure drop at flooding velocity 
x_pressure_drop=3600*void_fraction**2+186480*h_dyno_0*void_fraction+32280*d_l*at*void_fraction+191844*h_dyno_0**2+95028*d_l*at*h_dyno_0+10609*d_l**2*at**2
print(x_pressure_drop)
delta_p_tot_flood=(rho_l*g)/(2988*h_dyno_0)*(249*h_dyno_0*(x_pressure_drop**0.5-60*void_fraction-558*h_dyno_0-103*d_l*at))**0.5
print("Pressure drop at flooding velocity is :",delta_p_tot_flood)




```
