```python
import math 
g=9.81
temp=input("Enter the temperature in degree celcius for air:")
t_air=float(temp)
T=273.15+t_air
den_g=10**(-5)*t_air**2-0.0044*t_air+1.2877                     #Kg/m3
visco_g=-9*pow(10,-8)*t_air**2+0.0002*t_air+0.0623                  #Pa.hr
Dg=diffu_g=5*pow(10,-7)*t_air**2+0.0005*t_air+0.0684               #m2/hr
at=330                                                            #packing density  m2/m3
#R=82.057*pow(10,-3)                                                 #universal gas constant m3.atm/Kg.mol.K
R=8314
desiccant_flow=input("Enter the Desiccant flow rate in Kg/m2.s :")     #Desiccant flow rate in Kg/m2.hr          #proper value need to be eneterd i guesss as in hour
ratio=input("Enter the L/G ratio :")                                  #Liquid to air flow rate 
ratio=float(ratio)
L=float(desiccant_flow)
L=L*3600           #To convert the flow of desicant to hourly basis
G=L/ratio
Dp=0.001                                                              # Equivalent diameter of flow channel in meter
Kg=5.23*(at*Dg/(T*R))*(G/(at*visco_g))**0.7*(visco_g/(den_g*Dg))**(1/3)*(at*Dp)**(-2.0)                #units need to be checked
print("The k type mass transfer coefficient of the air is Kg:",Kg)
#P=total pressure here units are in atm so P=1atm and hence the Fg=Kg
P=1                                                                  # Total Presssure in atmospheric pressure
Fg=Kg*P                                                                #to convert the K type coefficient in F type coefficient
print("the mass transfer coeffient:" ,Fg)                                    
#To calculate the density of LiCl solution
import math
c1=1.9937718430
c2=1.0985211604
c3=-0.5094492996
c4=-1.7619124270
c5=-44.9005480267
c6=-723692.2618632
temp_desiccant=input("Enter the desiccant inlet tenperature in degree celcius:")
t_des=float(temp_desiccant)
thetha3=(t_des+273.15)/647.3
tau=1-thetha3
conc3=input("Enter the Concentration of the solution:")
x=float(conc3)
y=(x/(1-x))
critical_density_of_water=322   # kg/m3
density_of_water=round(critical_density_of_water*(1+c1*tau**(1/3)+c2*tau**(2/3)+c3*tau**(5/3)+c4*tau**(16/3)+c5*tau**(43/3)+c6*tau**(110/3)),4)
# print ("The density of the water at given temperature is {}" .format(density_of_water))
f1=1.0
f2=0.540966
f3=-0.303792
f4=0.100791
density_of_solution=round(density_of_water*(f1+f2*y**2+f3*y**3+f4*y**4),4)
print ("the density of the LiCl soltution at {} celcius, {} concentration is {} Kg/M3 ".format(t_des,x,density_of_solution))
rho_l=density_of_solution
#print(rho_l)
#To calculate the thermal conductivity of LiCl solution
alpha1=-11.7882*10**-3
alpha0= 10.8958*10**-3
alphaR=alpha0+x*alpha1
thermal_conducti_water20=0.6075 # W/m.K
thermal_conducti_water= thermal_conducti_water20*(0.208495+1.747278*thetha3)
# print("the thermal conducticity of pure water at given temperature: {} W/m.K ".format(thermal_conducti_water))
Is=1                         #mostly equal to 1 for diluted solutions
M=18   #Molar mass of water
equi_condu_depression= x*rho_l*Is/M
ther_condu_solu =thermal_conducti_water-equi_condu_depression*alphaR
print("The thermal conductivity of the solution is {} W/m.K".format(ther_condu_solu))
#To calculate the specefic heat capacity of the solution
e1=1.43980
e2=-1.24317
e3=-0.12070
e4=0.12825
e5=0.62934
e6=58.5225
e7=-105.6343
e8=47.7948
peta=((t_des+273.15)/228)-1
cp_water_temp=(2*10**-9*t_des**4-4*10**-7*t_des**3+5*10**-5*t_des**2-0.0021*t_des+4.2104)*1000  #multiplied by 1000 so converted to J/Kg.K
#print(cp_water_temp)
if x<=0.31 :
  e9=e1*x+e2*x**2+e3*x**3
else :
  e9=e4+e5*x
e10=e6*peta**0.02+e7*peta**0.04+e8*peta**0.06
#print(e10)
#print(e9)
cp_sol_peta=cp_water_temp*(1-e9*e10)
print("The specific thermal capacity of the solution is {} j/Kg.K ".format(cp_sol_peta))        #j/Kg.K
#To calculate the viscosity of the solution
#dynamic density of licl
c1=1.0261862
c2=12481.702
c3=-19510.923
c4=7065.286
c5=-395.561
c6=143922.996
visco_water_0celci=0.001793 # Pa.second
thetha_visco=(t_des/228)-1
#print(thetha_visco)
visco_water_temp=(2*10**-11*t_des**4-6*10**-9*t_des**3+7*10**-7*t_des**2-5*10**-5*t_des+0.0017)*3600      # Pa.s converted to Pa.hr
#print(visco_water_temp)
d1=0.090481
d2=1.390262
d3=0.675875
d4=-0.583517
falta=(x/(1-x)**(1/0.6))
#print(falta)
visco_solu= visco_water_temp * (math.exp((d1*falta**3.6+d2*falta+d3*(falta/thetha_visco)+d4*falta**2)))      #Pa.s converted to Pa.hr
print ('The value of the solution viscosity is {} in Pa.s'.format(visco_solu))
#To calculate the surface tension of solution
surface_tension_0water=235.8  # mN/m
zeta= -0.625
mu= 1.256
theta2=((t_des+273.15)/647.3)
surface_tension_water=surface_tension_0water*(1-zeta*(1-theta2))*((1-theta2)**mu)
#print('the value of the surface tension of water at that temperature is {}  mN/m'.format(surface_tension_water))
b1=2.757115
b2=-12.011299
b3=14.751818
b4=2.443204
b5=-3.147739
surface_tension_solution= surface_tension_water * (1+(b1*x)+(b2*theta2*x)+(b3*x*theta2**2)+(b4*x**2)+(b5*x**3))  #mN/m
print('the value of the surface tension of solution at that temperature is {}  mN/m'.format(surface_tension_solution))
#print(surface_tension_solution/surface_tension_water) 
#To calculate the liquid diffusisvity=D_l
D_l=ther_condu_solu/(rho_l*cp_sol_peta)                                          #Themal diffusivity od liquid in m2/hr                        
#To calculate the desiccant mass transfer coefficient
Kl=(visco_solu*g/rho_l)**(1/3)*0.0051*(L/(at*visco_solu))**(2/3)*(visco_solu/(rho_l*D_l))**(-1/2)*(at/Dp)**0.4    # final unit coming tobe(hr/m^0.4)^1/3
print("The liquid K type mass transfer coefficient is: {}".format(Kl))
# to convert k type coefficent to f type coefficient 
Xsm=1  #for dilute solution
Fl=Kl*Xsm*rho_l/M
print("The Fl is for desiccant is : {}".format( Fl))
# To calculate the vapor pressure of the solution
import numpy as np 
T=273.15+t_des
z=T/647.096
Z1=np.exp((1/z)*(-7.85951783*(1-z)+1.84408259*(1-z)**1.5-11.7866497*(1-z)**3+22.6807411*(1-z)**3.5-15.9618719*(1-z)**4+1.86122502*(1-z)**7.5))
ps_water_temp=22064*Z1  #in Kpa water saturation pressure
print("Sat pres of water is: {} Kpa" .format(ps_water_temp))
z0=0.28
z1=4.30
z2=0.60
z3=0.21
z4=5.10
z5=0.49
z6=0.362
z7=-4.75
z8=-0.40
z9=0.03
M=2-(1+(x/z0)**z1)**z2
N=(1+(x/z3)**z4)**z5-1
z10=M+N*z
z11=1-(1+(x/z6)**z7)**z8-z9*np.exp((x-0.1)**2/(-0.005))
pressure_solution_conc_temp=z10*z11*ps_water_temp      #Kpa
print("The vapor pressure of water vapor in solution is {} Kpa" .format(pressure_solution_conc_temp))
#to calculate the change in humidity of the system
dY_N=input("Enter the inlet specific moisture in Kg/Kg of dry air :")
dY_N=float(dY_N)     #Kg/Kg of dry air
H=input("Enter the height of the column in meter:")
H=float(H)      # meter
n=input("Enter the number of the steps in the column:")
n=float(n)
dz=H/n
MW=18
P=101.325 # atm to Kpa
moist1=dY_N1=dz*(dY_N/dz-(MW*Fg*at/G)*math.log((1-(pressure_solution_conc_temp/P))/(1-(ps_water_temp/P))))
print("%%The moisture output at the end of the step size {} is {} Kg/Kg of dry air".format(dz,moist1))

slope=dY_N1/dz
dL_N=L
dL_N1=dz*(dL_N/dz+G*slope)
print("%%The outlet mass flow rate of desiccant at outlet is {} kg/m2.s" .format(dL_N1/3600))

dX_N=x
dX_N1=dz*(dX_N/dz-G*x*slope/L)
print("%%THE OUTLET CONCENTRAION (weight fraction) OF THE DESICCANT SOULTION IS {}".format(dX_N1))

#TO CALCULATE THE SPECIFIC HEAT OF THE WATER VAPOR 
Cpv=0.016*t_air**2+0.1541*t_air+1858.7      #in j/Kg.K and temp in celci
print("The  SPECIFIC HEAT OF THE WATER VAPOR is {} j/Kg.K ".format(Cpv))

#TO CALCULATE THE SPECIFIC HEAT OF THE AIR
Cpa=-5*10**(-7)*t_air**4+7*10**(-5)*t_air**3-0.0019*t_air**2+0.0044*t_air+1006.2      #Temp in celci and CPa in J/Kg.K
print("The  SPECIFIC HEAT OF THE AIR is {} j/Kg.K ".format(Cpa))

#To CALCULATE THE THERMAL CONDUCTIVITY OD AIR WITH TEMP
Ka=-3*10**(-8)*t_air**2+8*10**(-5)*t_air+0.0244    #In W/m.k with celci of air
print("The Thermal Conductivity of air is {}  ".format(Ka))

#To calculate the smidth Number
Sc=visco_g/(den_g*Dg)
print("The Smidth number is {}  ".format(Sc))

#To Calculate the prandtl number
Pr=(Cpa*visco_g/(Ka*3600))
print("The Prandtl number is {}  ".format(Pr))

MA=28.97   #The molecular weight of dry air is 28.97 grams per mole

#TO CALCULATE THE HEAT TRANSFER COEFFIECNT FOR AIR SIDE hg
h_g=Fg*MA*(Cpa+dY_N*Cpv)*(Sc**(2/3))/(Pr**(2/3))
print("THE HEAT TRANSFER COEFFIECNT FOR AIR SIDE is {} ".format(h_g))

h_g_desh_into_at=-G*Cpv*slope/(1-math.exp(G*Cpv*slope/(h_g*at)))

#To calculate the outlet temp of air dTa_N1
dTa_N=t_air
dTa_N1=dz*(dTa_N/dz+(h_g_desh_into_at*(t_des-t_air))/(G*(Cpa+dY_N*Cpv)))
print("%%THE OUTLET TEMPERATURE OF AIR AT THE EXIT IS {} celcius".format(dTa_N1))

#TO CALCULATE THE DESICCANT OUTLET TEMP 
slope2=(h_g_desh_into_at*(t_des-t_air))/(G*(Cpa+dY_N*Cpv))
dTL_N=t_des
T_R=0   #reference temeprature here 0 degree celci 
lambda_R=2230 # latent heat of condensation in 2230j/gm at T_R which is at 0 degree celci
dTL_N1=dz*((dTL_N/dz)+(G/(cp_sol_peta*L))*((Cpa+dY_N*Cpv)*slope+(Cpv*(t_air-T_R)-cp_sol_peta*(t_des-T_R)+lambda_R)*slope2))
print("%%THE OUTLET TEMPERATURE OF desiccant AT THE EXIT IS {} celcius".format(dTL_N1))
```
